# 2. Factors affecting lift and drag

## 2.1 Action and Reacting from Air
To generate supporting force, a mass of air must be accelerated or deflected to yield an upward reaction which for equilibrium, equals weight. To work on the air, the wing or wings of the model must move through it, disturbing it. __The greater the expenditure of energy required to generate lift, the less the efficiency of the model.__

The mass of the air available for a model to work depends on three factors:
1. density ($\rho$)
2. the size of the model
3. the speed of flight

## 2.2 Density
Density is the measure of the number of molecules in a given space. Model aircraft do not fly at such speeds that the compressibility of the air needs to be allowed for. This is true also for full-sized sailplanes and hang gliders, ultra light, light and commercial aircraft up to medium sized piston-engined transport aeroplanes.

__At high altitudes and in hot weather, the air is less dense near sea level when cold.__

In aerodynamics, a standard value for density of $1.225 kg/m^3$ or $0.002378 slugs/ft^3$ ($0.07651 lbm/ft^3$) for air corresponding to sea level at normal temperature and pressure.

## 2.3 Model Size
A large model flying through air must create more disturbance and hence generate more air reaction, both lift and drag, than a small model at similar speed. Span loading is expressed as a ratio, weight-per-unit length (Newtons per meter of pounds per feet). The capital __W__ stands for weight and the lowercase __b__ stands for span, thus: $$\mathbf{Span \,Loading = \frac{W}{b}}$$

Model size if more conviniently expressed in terms of wing area. The capital __S__ is used to stand for area.

## 2.5 Angle of Attack and Trim
However large and fast a model may be, it's ability to gain lift will depend almost entirely on the form of the wing and its _angle of attack_ relative to the airflow. The angle of attack is measured in degrees from some more or less arbitrary reference line, usually the straight line through the extreme leading and trailing edges of the wing airfoil section or profile.

The tailplane is a small wing which may or may not contribute lift to the total, but whose main function is to trim the mainplane to the desired angle of attack and hold it there. _Angle of incidence_ refers only to the rigging angle of such surfaces relative to some datum line on the drawing board. Strictly, almost all ordinaty airplanes and gliders are __tandems__ in that they have two wing-like surfaces disposed one behind the other and set at different rigging angles relative to one another. If one of the pair of wings carries no load or very little, it functions only as a stabilizer and control surface and may be very small relative to the main load-carrying wing.

## 2.6 Airfoil Sections and Lift Coefficients
The efficiency of a wing is influenced by its airfoil section which has some degree and type of __camber__ and some __thickness__ form.

<img src=http://www.aerospaceweb.org/question/airfoils/airfoil/airfoil-parts.jpg align="center"/>

For convinience, aerodynamicists adopt a convention which allows all the very complex factors of wing trim and shape to be summed up in one figure, the __coefficient of lift__ ($C_L$). An increase in of density, wing area, speed or $C_L$ will increase lift generation, thus: $$L = f(\rho, V, S, C_L)$$

The standard formula for lift was developed by Daniel Bernoulli:
$$\mathbf{L = \frac{1}{2} \,\rho \,V^2 \,S \,C_L}$$

Since V is squared, a relative small increase in V yields a large increase in lift.

## 2.7 Wing Loading
The wing loading, often written $\mathbf{W \,/ \,S}$ and expressed as kilograms per square meters, is the easiest way of potraying this relationship.
$$\mathbf{\frac{W}{S} = \frac{L}{S} = \frac{\rho V^2 C_L}{2}}$$\
For gliders, and descending power models, lift and weight are not quite equal ($L = W cos(\alpha)$) but for normal angles of dive or climb less than ten degress there is very little difference and the wing loading formula holds good.

## 2.8 Wing $C_L$ and Section $c_l$\
The $C_L$ of a whole model or whole wing should not be confused with the lift coefficients determined in a wind tunnel for an airfoil section. The section lift coefficient is often expressed as $c_l$.

## 2.9 Wing $C_L$ and Total $C_L$
If the tail or canard forewing does contribute some lift to the total $C_L$ of the whole model may be determined using the combined areas in the lift formula. In full-sized aeronautics, the aircraft $C_L$ is usually determined in terms of the wing area alone.

## 2.10 Streamlined Flow
If either the angle of attack or the camber is increased too much, the streamlining breaks down and the flow separates from the wing. Flow separation not only creates a great deal of drag, but also changes markedly the pressure difference between upper and lower surfaces. The lift force is drastically reduced, the wing is stalled.

## 2.12 Bernoulli's Theorem
Bernoulli's theorem connects the pressure measured at any point in a fluid such as air to the mass density and velocity of the flow. if there is a pressure difference in any sense, the particle will accelerate or decelerate in accordance with the second law of motion. Velocity will increase if the pressure on the front face of the cylinder is less than that behind. Velocity will decrease if the pressure behind is less than in front. Hence the particle will speed up as it approaches a region of low pressure, and slow down on approaching a high pressure zone. Since it is not isolated, but part of a general streamlined flow, the same laws apply to every particle in the flow, which therefore speeds up and slows down on approaching low and high pressure regions respectively.

$$\mathbf{P + \frac{\rho \, V^2}{2} = Constant} \;\;\;(Bernoulli's Theorem)$$

Pressure and velocity are the only variable; if one increases, the other decreases under all circumstances.

## 2.13 Drag, the Lift: Drag Ration
All parts of a model, including wings, tail fuselage and every component exposed to the air flow, contribute drag.

$$\mathbf{D = \frac{\rho \, V^2 \, S \, C_D}{2}}$$

The S, or area in this formula is normally the wing area of the whole aircraft. If the total surface area is used (including tail) for the $C_L$, the same total area must be used for the drag equation. This enables the drag and lift forces to be compared, usually in the form of a ratio called the __lift to drag ratio__ or $L \, / \, D$.

At some speed, the model achieves its maximum $L \, / \, D$ ratio. The value of this ratio gives a rough measure of the all-around efficiency of the model. Modellers quite often speak of increasing lift by changing the trim or using a different wing section. If the drag is regarded as the inevitable price paid for keeping a given model in the air, reducing the drag price always makes for a more efficient flight.

## 2.14 Vortex Drag
__Induced drag__ is now called _vortex_ drag because it is associated with the rotating vortices which trail behind any wing, or any surface, which is yielding aerodynamic lift. The induced drag increases as the velocity decreases.

$$Induced \, Drag \; \alpha \; \frac{L}{V^2}$$

## 2.15 Profile Drag
Form or pressure drag is caused by the total of all the pressure variations over a body as the air flows round it, and skin frictionor viscous drag is caused by the contact of the air with the model's surfaces. Although it is useful to separate these different types of drag for purposes of study, it is clear that they almost always occur together.

For this reason, particularly, when wings are concerned, skin friction and form drag are commonly taken together and termed _profile drag_.

## 2.16 Total Drag
The total drag of the aircraft is composed of the total vortex and all other drags at each speed. Where the vortex draf equals all the rest, i.e., there the two lower curves intersect, drag is a minimum for the whole aircraft. Since lift is constant, for a given mass of aircraft in level flight, it is at the minimum drag point on the curve that the best $L \, / \, D$, as _sinking speed_ plotted agains velocity rather than total drag agains $C_L$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Velocity
velocity  = np.linspace(0.3, 2, 1000)

# Induced Drag
ind_drag = 1 / (velocity ** 2)

# Profile Drag
prof_drag = velocity ** 2

# Total Drag
total_drag = ind_drag + prof_drag

# Plot drag data
plt.figure()
plt.grid(True, linestyle=':')

plt.plot(velocity, ind_drag)
plt.plot(velocity, prof_drag)
plt.plot(velocity, total_drag)

plt.xlabel("Velocity", fontweight="bold")
plt.ylabel("Drag", fontweight="bold")

plt.legend(["Induced Drag", "Profile Drag", "Total Drag"])

plt.show()
```
