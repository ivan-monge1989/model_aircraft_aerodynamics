# 1. Fundamentals

## 1.1 Laws of Motion
All aerodynamic theory depends on the laws of motion. These were originally worked out by Isaac Newton, remain entirely valid in engineering providing that the matters under discussion are of "ordinary" size and densities and the velocity remains well below the speed of light.

## 1.2 Equilibrium
If a body is in equilibrium, it tend to remain so. All the forces acting on an object in equilibrium are in balance. There is no tendency for it to change its state or accelerate, decelarate or change direction. Such bodies stay put unless something disturbs them. A model flying straight and level in calm air, neither speeding up or slowing down, nor turning, is in a balanced stay and will tend to continue moving steadily. Equilibrium, then, is a very common state of affairs. It is a condition of steady motion or rest, in contrast to states of unsteady motion involving acceleration and negative acceleration or deceleration.

## 1.3 Acceleration, Mass and Force
To disturb equilibrium, changing the speed or direction of flight in any way, requires a force to bring about an acceleration in the appropriate sense. The __second law of motion__ states tha the stregth of the force required for any given acceleration depends on the _mass_ of the model, i.e, $$\mathbf{F \; \alpha \; m}$$

An object of large mass requires greater forces to disturb its equilibrium to any given extent than a small mass. But the larget mass also requires a large force to accelerate it to flying speed from standstill, more force to change the level of climb, etc. Whenever there is a disturbance of equilibrium, _inertia_ oposes the change.

## 1.4 Action and Reaction
The __third law of motion__ establishes that action and reaction are equal and opposite. In level flight, the _weight force_ (__W__) acts vertically downwards. The force acting in the upwards direction, for level flight, is called _lift_ (__L__). Lift is generally generated by the wings of the plane and to some minor extend some can be generated by the tailplane. A helicopter or drone generates lift by its rotating propellers.

## 1.5 Resolution of Forces
The main upward support comes from the wings, but the taiplane also may provide some lift, so its contribution must be added to the total vertical reaction. 
* L = lift
* W = weight
* D = drag
* T = thrust
* $\alpha$ = Angle of Attack (AoA)

### 1.5.1 Level Flight Resolution of Forces
__X - Direction:__    $$\;\;\;L_{total} = W_{total}$$

__Y - Direction:__    $$\;\;\;D_{total} = T_{total}$$

In level flight, all forces are balanced.

## 1.5.2 Gliding (No power) Resolution of Forces
__X - Direction:__    $$\;\;\;L = W cos(\alpha)$$

__Y - Direction:__    $$\;\;\;D = W sin(\alpha)$$

The weight alone acts vertically downwards, but may be resolved into one force acting along the flight path and another at right angles to it. The glider moves forward and slightly downwards under the action of the weight component along the flight direction. The total air reaction force is similarly resolved into lift at right angles to the flight and drag opposing the forward-acting force.

## Climbing Resolution of Forces
__X - Direction:__    $$\;\;\;L = W cos(\alpha)$$

__Y - Direction:__    $$\;\;\;T = W sin(\alpha) + D$$

In a climb, the total support comes from a combination of wings and propeller. The weight may be resolved into two components, one opposing lift and the other directly opposing thrust, assisting the drag.
